It seems there should be columns called Order Id, Segment containing values like Consumer, Corporate and Home office and Ship Mode covering information like First class, Same day, Second class, Standard class and another column showing sales value.

```python
#Since first two rows and last row dont keep main information, I delete them
data.drop(index=data.index[0:2], axis=0, inplace=True)
data.drop(index=data.index[-1], axis=0, inplace=True)
```

```python
#reset index
data.reset_index(drop=True, inplace=True)
```

```python
#Renaming the column Order_Id
data.rename(columns={"Segment>>": "Order_Id"}, inplace=True)
```

```python
#Check again size od the dataset
data.shape
```

```python
#Check missing values again
data.isnull().sum()
```

```python
# Create the column Segment with null values
data["Segment"]=np.nan
```

```python
# The indexes of Segment will be filled with the value "Consumer" where the indexes of the column Consumer, Unnamed:2, Unnamed:3, Unnamed:4 are not null.
row_consumer= data.dropna(subset=['Consumer'], axis=0).index
data.loc[row_consumer, "Segment"]="Consumer"

row_un_2= data.dropna(subset=['Unnamed: 2'], axis=0).index
data.loc[row_un_2, "Segment"]="Consumer"

row_un_3= data.dropna(subset=['Unnamed: 3'], axis=0).index
data.loc[row_un_3, "Segment"]="Consumer"

row_un_4= data.dropna(subset=['Unnamed: 4'], axis=0).index
data.loc[row_un_4, "Segment"]="Consumer"
```

```python
# The indexes of Segment will be filled with the value "Corporate" where the indexes of the column Corporate, Unnamed:7, Unnamed:8, Unnamed:9 are not null.
row_cor_= data.dropna(subset=['Corporate'], axis=0).index
data.loc[row_cor_, "Segment"]="Corporate"

row_cor_7= data.dropna(subset=['Unnamed: 7'], axis=0).index
data.loc[row_cor_7, "Segment"]="Corporate"

row_cor_8= data.dropna(subset=['Unnamed: 8'], axis=0).index
data.loc[row_cor_8, "Segment"]="Corporate"

row_cor_9= data.dropna(subset=['Unnamed: 9'], axis=0).index
data.loc[row_cor_9, "Segment"]="Corporate"
```

```python
# The indexes of Segment will be filled with the value "Home Office" where the indexes of the column Corporate, Unnamed:12, Unnamed:13, Unnamed:14 are not null.
row_ho_= data.dropna(subset=['Home Office'], axis=0).index
data.loc[row_ho_, "Segment"]="Home Office"

row_ho_12= data.dropna(subset=['Unnamed: 12'], axis=0).index
data.loc[row_ho_12, "Segment"]="Home Office"

row_ho_13= data.dropna(subset=['Unnamed: 13'], axis=0).index
data.loc[row_ho_13, "Segment"]="Home Office"

row_ho_14= data.dropna(subset=['Unnamed: 14'], axis=0).index
data.loc[row_ho_14, "Segment"]="Home Office"
```

```python
#Create Sale column with null values
data["Sale"]=np.nan
```

```python
# The indexes of Sale will be filled with the same value which is not null in the indexes of Consumer Total 
data["Sale"]=np.where(~data["Consumer Total"].isnull(), data["Consumer Total"], data["Sale"])
```

```python
# The indexes of Sale will be filled with the same value which is not null in the indexes of Corporate Total 
data["Sale"]=np.where(~data["Corporate Total"].isnull(), data["Corporate Total"], data["Sale"])
```

```python
# The indexes of Sale will be filled with the same value which is not null in the indexes of Home Office Total 
data["Sale"]=np.where(~data["Home Office Total"].isnull(), data["Home Office Total"], data["Sale"])
```

```python
data["Sale"].isnull().sum()
```

```python
#Create Ship_mode column with null values
data["Ship_mode"]=np.nan
```

```python
# The indexes of Ship_mode will be filled with the value "First class, Same day, Second Class, Standard Class" in accordance with the indexes of the column Consumer, Unnamed:2, Unnamed:3, Unnamed:4 are not null.

data["Ship_mode"]=np.where(~data["Consumer"].isnull(), "First Class", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 2"].isnull(), "Same day", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 3"].isnull(), "Second Class", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 4"].isnull(), "Standard Class", data["Ship_mode"])
```

```python
# The indexes of Ship_mode will be filled with the value "First class, Same day, Second Class, Standard Class" in accordance with the indexes of the column Corporate, Unnamed:7, Unnamed:8, Unnamed:9 are not null.

data["Ship_mode"]=np.where(~data["Corporate"].isnull(), "First Class", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 7"].isnull(), "Same day", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 8"].isnull(), "Second Class", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 9"].isnull(), "Standard Class", data["Ship_mode"])
```

```python
# The indexes of Ship_mode will be filled with the value "First class, Same day, Second Class, Standard Class" in accordance with the indexes of the column Home Office, Unnamed:12, Unnamed:13, Unnamed:14 are not null.

data["Ship_mode"]=np.where(~data["Home Office"].isnull(), "First Class", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 12"].isnull(), "Same day", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 13"].isnull(), "Second Class", data["Ship_mode"])

data["Ship_mode"]=np.where(~data["Unnamed: 14"].isnull(), "Standard Class", data["Ship_mode"])
```
